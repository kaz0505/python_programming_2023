## 5. AIライブラリの確認
- AIでよくつかわれる sklearnライブラリの動作を確認します。
- IRISデータセットを、SVCアルゴリズムを使って学習し、いくつかのデータの判別を行います。

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import datasets
from sklearn.metrics import accuracy_score

# データセットの読み込み
iris = datasets.load_iris()
x, y = iris.data, iris.target

# 教師データとテストデータに分ける
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)

# モデルとしてサポートベクターマシンを使う
model = svm.SVC()

# 教師データを使って学習する
model.fit(x_train, y_train)

# 学修済みモデルを使って、テストデータを評価する
pred = model.predict(x_test)
print( "正解率 {:.4g} %".format(accuracy_score(y_test, pred)*100))

# 学習済みモデルを使って、データを判別する
data = [ [1.4, 3.5, 5.1, 0.2], [6.5, 2.6, 4.4, 1.4], [5.9, 3.0, 5.2, 1.5] ]
for item in data:
  print("データ {} -> 判別結果 {}".format(item, model.predict([item])))
```
